```python
#step 1: import the below libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#step 2: path_of_the_folder_csvFile_saved
mydir="C:/Users/gande/OneDrive/Desktop/A/project/Amazon--py/"
```

```python
# Initialize an empty list to store dataframes
mylist=[]
```

```python
# Loop through files in the directory and read CSV files into pandas dataframes
for file in os.listdir(mydir):
    if file.endswith('.csv'):
        data=pd.read_csv(mydir+file)
        mylist.append(data)
    else:
        continue

data=pd.concat(mylist)
```

```python
# Display the column names of the dataframe 'data'
data
```

```python
# Display information about the dataframe
data.info()
```

```python
# Drop the column 'Unnamed: 0' from the dataframe 'data'
data.columns
```

```python
# Drop the column 'Unnamed: 0' from the dataframe 'data'
data = data.drop(columns=['Unnamed: 0'])
```

```python
# Convert 'ratings' column to numeric type
data['ratings'] = pd.to_numeric(data['ratings'], errors='coerce')

# Convert 'no_of_ratings' column to integer type with nullable integers
data['no_of_ratings'] = pd.to_numeric(data['no_of_ratings'], errors='coerce').astype('Int64')

# Clean 'discount_price' and 'actual_price' columns by removing currency symbols and converting to float
data['discount_price'] = data['discount_price'].str.replace('₹', '').str.replace(',', '').astype(float)
data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)

```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['ratings']=data['ratings'].fillna(data['ratings'].median())
```

```python
data['no_of_ratings'] = data['no_of_ratings'].fillna(0)
```

```python
data['discount_price'] = data['discount_price'].fillna(0)
```

```python
data['actual_price'] = data['actual_price'].fillna(0)
```

```python
data.info()
```

```python
subcategory_counts = data.groupby('main_category')['sub_category'].nunique()

print(subcategory_counts)

```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
subcategory_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Subcategories per Main Category')
plt.xlabel('Main Category')
plt.ylabel('Number of Subcategories')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Iterate through unique combinations and print sub_category for each main_category
for main_category in data['main_category'].unique():
    print(f'{main_category}:')
    for sub_category in data[data['main_category'] == main_category]['sub_category'].unique():
        print(f'  {sub_category}')
    print()
```

```python
rating_counts = data.groupby('main_category')['ratings'].count()

# Plotting the data
plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Ratings per Main Category')
plt.xlabel('Main Category')
plt.ylabel('Count of Ratings')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

# Plotting histogram of ratings
plt.hist(data['ratings'], bins=30, edgecolor='black')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Histogram of Ratings')
plt.show()
```

```python
Actual_prices = data.groupby('main_category')['actual_price'].sum()

# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(Actual_prices, labels=Actual_prices.index, autopct='%1.1f%%', startangle=20)
plt.title('Distribution of Actual Prices by Main Category')
plt.axis('equal')  
plt.show()

```

```python

# Sort values in descending order and select top 5
top_categories = Actual_prices.sort_values(ascending=False).head(5)

for category, price in top_categories.items():
    print(f'{category}: {price:.0f}')
```

```python
# Group by main_category and sum the actual_price
Actual_prices = data.groupby('main_category')['actual_price'].sum()

for category, price in Actual_prices.items():
    print(f'{category}: {price:.0f}')
```

```python
# Calculate sum of actual_price grouped by main_category and sub_category
category_sub_prices = data.groupby(['main_category', 'sub_category'])['actual_price'].sum()

# Function to get top 5 sub-categories for each main category
top_categories = category_sub_prices.groupby(level=0, group_keys=False).apply(lambda group: group.nlargest(5))

# Print top 5 sub-categories for each main category
printed_main_categories = set()
for (main_category, sub_category), price in top_categories.items():
    if main_category not in printed_main_categories:
        print(f'{main_category}:')
        printed_main_categories.add(main_category)
    print(f'  {sub_category}: {price:.0f}')
```

```python
# Group by main_category and sum the actual_price
discount_price = data.groupby('main_category')['discount_price'].sum()

# Print exact values for each main_category
for category, price in discount_price.items():
    print(f'{category}: {price:.0f}')
```

```python
# Plotting pie chart
plt.figure(figsize=(8, 6))
plt.pie(discount_price, labels=discount_price.index, autopct='%1.1f%%', startangle=20)
plt.title('Distribution of Discount_price by Main Category')
plt.axis('equal')  
plt.show()
```
